```python
icici = pd.read_csv("H:\\MSIT 2nd Year\\DADV\\Selenium\\Analysis\\BSE\\ICICI.csv")["Close Price"].reset_index()
```

```python
df = pd.DataFrame()
newDf = pd.DataFrame()
for i in range(10):
    link = "C:\\Users\\Someshwar Thumma\\Downloads\\"+str(finalList[i])+".csv"
    data = pd.read_csv(link)["Close Price"].reset_index()
    df = pd.merge(data,icici,on='index').iloc[:,[1,2]]
    newRow = {'Name':securityName[i], 'Corr':df.corr()['Close Price_x'][1]}
    newDf = newDf.append(newRow, ignore_index=True)
```

```python
newDf.sort_values("Corr", axis = 0, ascending = False, inplace = True, na_position ='last')
str(newDf['Name'][0])+ " has the highest correlation with ICIC Bank with a correlation of " +str(newDf['Corr'][0])
```
