```python
import pandas as pd
import numpy as np
import time
```

```python
while True:
    try:
        basics = pd.read_csv('H:\\MSIT 2nd Year\\DADV\\Selenium\\IMDB Data\\Data\\basics.tsv',delimiter='\t',encoding='utf-8')
        break
    except:
        print(" not reading..")
        time.sleep(5)
        pass
```

```python
while True:
    try:
        ratings = pd.read_csv('H:\\MSIT 2nd Year\\DADV\\Selenium\\IMDB Data\\Data\\ratings.tsv',delimiter='\t',encoding='utf-8')
        break
    except:
        print(" not reading.. just waiting")
        time.sleep(5)
        pass
```

```python
rating_basics = pd.merge(basics,ratings,on='tconst')
rating_basics.head()
```

```python
rating_basics.sort_values("averageRating", axis = 0, ascending = False, inplace = True, na_position ='last')
rating_basics['originalTitle'].iloc[:20]
```

```python
rating_basics["startYear"] = pd.to_numeric(rating_basics["startYear"], errors='coerce')
```

```python
tempDf = rating_basics[rating_basics.startYear>2000]
tempDf = tempDf[tempDf.numVotes>40000]
tempDf["popular"] = tempDf.numVotes*tempDf.averageRating
tempDf.head()
```

```python
tempDf.sort_values("popular", axis = 0, ascending = False, inplace = True, na_position ='last')
print(list(tempDf.originalTitle.head(20)))
```
